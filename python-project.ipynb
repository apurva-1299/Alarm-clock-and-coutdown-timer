```python
#Importing all the necessary libraries to form the alarm clock:
from tkinter import *
import datetime
import time
from playsound import playsound
import pygame
from tkinter import ttk
import tkinter.font as font
```

```python
def SubmitButton():
    AlarmTime= entry1.get()
    Message1()
    CurrentTime = time.strftime("%H:%M")
    print("the alarm time is: {}".format(AlarmTime))
      #label2.config(text="")
    while AlarmTime != CurrentTime:
        CurrentTime = time.strftime("%H:%M")
        time.sleep(1)
    if AlarmTime == CurrentTime:
        print("now Alarm Musing Playing")
        pygame.mixer.init()
        pygame.mixer.music.load('C:\\Users\\HP\\Desktop\\ocr\\python\\alarm-music.mp3')
        pygame.mixer.music.play()
        label2.config(text = "Alarm music playing.....",font = ('Helvetica', 10, "bold"))
        messagebox.showinfo(title= 'Alarm Message', message= "{}".format(entry2.get()))
def Message1():
    AlarmTimeLable= entry1.get()
    label2.config(text="the Alarm time is Counting...",font = ('Helvetica', 10, "bold"))
    messagebox.showinfo(title = 'Alarm clock', message = 'Alarm will Ring at {}'.format(AlarmTimeLable))

def clear():
    entry1.delete(0,"end")
    entry2.delete(0,"end")
    label2.pack_forget()
    pygame.mixer.music.pause()

def clear2():
    pygame.mixer.music.pause()

def tick(val):
    global clock
    window=Toplevel()
    window.geometry("300x130")
    clock=Label(window, font = ('Helvetica', 30, "bold"), fg="red", bg="black", bd=16, relief=SUNKEN)
    clock.place(x=18, y=30)
    if val=="btn_24":
        time24()
    elif val=="btn_12":
        time12()
    else:
        alarm()

def time24():
    global clock
    ts=time.strftime("%H:%M:%S")
    clock["text"]=ts
    clock.after(1000, time24)

def time12():
    global clock
    ts=time.strftime("%I:%M:%S:%p")
    clock["text"]=ts
    clock.after(1000, time12)

def submit1():
    try:
        temp = int(hour.get())*3600 + int(minute.get())*60 + int(second.get())
    except:
        print("Please input the right value")
    while temp >-1:
         
        # divmod(firstvalue = temp//60, secondvalue = temp%60)
        mins,secs = divmod(temp,60) 
  
        # Converting the input entered in mins or secs to hours,
        # mins ,secs(input = 110 min --> 120*60 = 6600 => 1hr :
        # 50min: 0sec)
        hours=0
        if mins >60:
             
            # divmod(firstvalue = temp//60, secondvalue 
            # = temp%60)
            hours, mins = divmod(mins, 60)
         
        # using format () method to store the value up to 
        # two decimal places
        hour.set("{0:2d}".format(hours))
        minute.set("{0:2d}".format(mins))
        second.set("{0:2d}".format(secs))
  
        # updating the GUI window after decrementing the
        # temp value every time
        root.update()
        time.sleep(1)
  
        # when temp value = 0; then a messagebox pop's up
        # with a message:"Time's up"
        if (temp == 0):
            pygame.mixer.init()
            pygame.mixer.music.load('C:\\Users\\HP\\Desktop\\ocr\\python\\alarm-music.mp3')
            pygame.mixer.music.play()
            messagebox.showinfo("Time Countdown", "Time's up ")
        # after every one sec the value of temp will be decremented
        # by one
        temp -= 1

def userText(event):
    entry1.delete(0,END)
    usercheck=True
```

```python
from tkinter import *
from tkinter import ttk
import time
import os
from tkinter import messagebox


root = Tk()
root.title("Alarm clock")
root.geometry("600x400")


style = ttk.Style()
style.configure("TNotebook.Tab", foreground="olive drab", borderwidth=100,font=('Helvetica','11','bold'))


style.configure('TButton', background="#8EA002", foreground="black")
style.configure('TLabel', foreground="black")

TAB_CONTROL = ttk.Notebook(root)

TAB1 = ttk.Frame(TAB_CONTROL, style = 'tab1.TFrame')
TAB_CONTROL.add(TAB1, text= ' Clock    ')

TAB2 = ttk.Frame(TAB_CONTROL, style = 'tab1.TFrame')
TAB_CONTROL.add(TAB2, text=' Set Alarm     ')

TAB3 = ttk.Frame(TAB_CONTROL, style = 'tab1.TFrame')
TAB_CONTROL.add(TAB3, text=' Set Timer     ')

TAB_CONTROL.pack(expand=1, fill="both")


#TAB1 CLOCK
img1 = PhotoImage(file=r'c1.png',master = root)
photoimage1 = img1.subsample(3, 3)
btn1 = ttk.Button(TAB1,image=photoimage1,style="TButton",command =lambda: tick("btn_12")).place(anchor=SW,x=65, y=133)
label1= Label(TAB1, text="12 Hour Clock",font = ('Helvetica', 19, "bold"),fg='orange red')
label1.pack(side= TOP, pady=60)

img2 = PhotoImage(file=r'c1.png',master = root)
photoimage2 = img1.subsample(3, 3)
btn2 = ttk.Button(TAB1,image=photoimage2,style="TButton",command =lambda: tick("btn_24")).place(anchor=SW,x=65, y=300)
label2= Label(TAB1, text="24 Hour Clock",font = ('Helvetica', 19, "bold"),fg='orange red')
label2.pack(side= TOP, pady=60)
#TAB1 END


#TAB2 SET ALARAM
usercheck=False
InputLabel= Label(TAB2,text = "Enter the Alarm time in 24 Hour Format :",font = ('Helvetica', 12, "bold"))
InputLabel.pack(side= TOP, pady=15)

entry1 = ttk.Entry(TAB2, width = 35)
entry1.pack()
entry1.insert(0,'example-13:45')
entry1.bind("<Button>",userText)

labelAlarmMessage= ttk.Label(TAB2, text="Alarm Message:",font = ('Helvetica', 12, "bold"))
labelAlarmMessage.pack(side= TOP, pady=15)

entry2= ttk.Entry(TAB2, width=35)
entry2.pack()

button1= Button(TAB2, text= "submit",bd='5', command=SubmitButton)
button1.pack(side= TOP, pady=20)

label2= Label(TAB2,fg='MediumOrchid1')
label2.pack()

ClearButton = Button(TAB2,text="Clear",bd='5',command = clear)
ClearButton.pack(side = TOP, pady=20,padx=15)
# TAB2 SET ALARAM END

# TAB3 TIMER 
hour=StringVar()
minute=StringVar()
second=StringVar()
hour.set("00")
minute.set("00")
second.set("00") 
label1= ttk.Label(TAB3, text="   Hour       Mins      Sec",font = ('Helvetica', 12, "bold"))
label1.pack(side= TOP, pady=15)
hourEntry= Entry(TAB3, width=3, font=("Arial",18,""),textvariable=hour)
hourEntry.place(x=230,y=40)
  
minuteEntry= Entry(TAB3, width=3, font=("Arial",18,""),textvariable=minute)
minuteEntry.place(x=290,y=40)
  
secondEntry= Entry(TAB3, width=3, font=("Arial",18,""),textvariable=second)
secondEntry.place(x=350,y=40)
  
btn = Button(TAB3, text='Set Time Countdown', bd='5',command= submit1)
btn.place(x = 245,y = 120)

ClearButton = Button(TAB3, text="  Stop ",bd='5',command = clear)
ClearButton.place(x=280,y=190)
# TAB3 TIMER END


root.mainloop()
```
